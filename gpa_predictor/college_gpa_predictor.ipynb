```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv("sat_gpa_v1.csv")
```

```python
x = df['sat_sum']
y = df['fy_gpa']
x = sm.add_constant(x)
```

```python
plt.rcParams["figure.figsize"] = [12, 6]
sns.regplot(data=df, x='sat_sum', y='fy_gpa', line_kws={"color": 'purple'}, scatter_kws={"alpha": 0.35})
plt.xlabel('SAT score')
plt.ylabel('Final year GPA')
plt.title('Correlation Between GPA & SAT Scores')
```

```python
OLS = sm.OLS(y,x).fit()
print(OLS.summary())
```

##### Explaining relationship and significance
    * Firstly, we would look at the p-value (P>|t|) for SAT scores. Typically in statistics if the p-value is less than 0.05 there is evidence of a statistically significant relationship between our 2 variables. In this model we get a value of 0.000, which means there is a strong correlation between SAT scores and final year GPA.
    * Secondly, we would look at the linear regression coefficient of SAT scores. Because the p-value is statistically significant and is typically assessed using the same rule, we can say that the linear regression coefficient, at 0.0024, also indicates a strong, positive correlation between SAT scores and college GPA. It is known that it is a positive correlation because the value is positive.
    * Finally, the F-statistic and its paired p-value give us more insight into the relationship between X and Y. Since our F-statistic is high (268.3) and its correlated p-value is very close to zero (1.39x10^-53), we can again conclude that there is a strong correlation between SAT scores and GPA.
